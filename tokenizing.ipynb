```python
#불용어 제거 함수 다시 생성
def rm_stop_words(noun_list):
    temp_list = [noun for noun in noun_list if not noun in stop_kr_list]
    return ','.join(temp_list)

no_sw_noun = df_noun['cont_noun'].apply(rm_stop_words)
```
