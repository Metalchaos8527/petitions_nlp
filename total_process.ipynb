```python
#전처리 시작1

#제목 이용 중복데이터 제거

#1) 시리즈를 리스트로
import re

raw_title = list(df_raw['title'])
raw_title[:10]
```

```python
raw_title[0].split()
```

```python
#제목변형 정규표현식
result = []

for text in raw_title:
    text_sp = text.split() #문자열들을 띄어씌기 기준으로 구분하여 리스트를 생성
    text_join = ''.join(text_sp) #리스트안의 문자열들을 하나의 문자열로 생성
    text_sub = re.sub(r'[,$()*+.?[\^{|]+', '', text_join) #문자열 안의 특수문자 제거
    result += [text_sub] #result 리스트안에 하나씩 변환된 제목들을 쌓기
```

```python
#한국 불용어 목록 불러오기
#다음의 페이지에서 참조함
#https://www.ranks.nl/stopwords/korean
#https://bab2min.tistory.com/544

stopwords_kr = pd.read_csv("C:/Users/myunghoon_k/OneDrive - 서울시립대학교/bitamin/NLP_summer/nlp_bitamin/stopwords_kr.txt",
                          sep = '\n', header = None)
stop_kr_list = list(stopwords_kr.iloc[:,0])
```

```python
#전체 텍스트 전처리 시도
#전처리하고자 하는 내용을 담은 사용자 정의 함수 실행
#https://regex101.com/ 

url_with_https = re.compile(r'(https?:\/\/){1,}([\w.\/%-]+){1,}')
rm_email = re.compile(r'([a-zA-Z0-9\.\_]+)@([\w\.]+){1,}')
useless_words = re.compile(r'[\(\)\"\'\∎\■\□\=\-\<\>\★\○]+')
rm_num = re.compile(r'([^\w]?\d+\.?\,?\)?\d*)+')

def clean_kor_text(kor_text):
    text = url_with_https.sub('',kor_text)
    text = rm_email.sub('', text)
    text = useless_words.sub('',text)
    text = rm_num.sub('',text)
    text = ' '.join(word for word in text.split() if word not in stop_kr_list)
    return text

df2['mod_content'] = df2['content'].apply(clean_kor_text)
```

```python
#명사태깅 실시
import numpy as np
import time

df3.reset_index(drop = True, inplace = True)
start_time = time.time()

result_cont = pd.Series(dtype = 'object') #빈시리즈 생성
for i in range(len(slice_list_final)-1):
    if i < 26:
        df3_f = df3.iloc[slice_list_final[i]:slice_list_final[i+1], 3]
        temp = df3_f.map(lambda x: get_nouns(x)) 
        result_cont = result_cont.append(temp, ignore_index = True)
    else:
        df3_f = df3.iloc[slice_list_final[i]:, 3]
        temp = df3_f.map(lambda x: get_nouns(x))
        result_cont = result_cont.append(temp, ignore_index = True)

print('소요시간: ', np.round((time.time() - start_time) / 60, 2))
result_cont.shape
    #약 3분이 소요됨!
```

```python
#TF-IDF 생성
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(max_features = 5000, max_df = 0.90, 
                     min_df = 5, sublinear_tf = True).fit_transform(df3['cont_noun_str'])
```

```python
tv_array = tv.toarray()
```

```python
#모델 학습

#1. 랜덤포레스트
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X = tv_array #tf-idf 행렬이 feature로 이용된다
y = df3.category #예측하고자 하는 것. 카테고리

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,
                                                   stratify = y, random_state = 156)
```

```python
#랜덤포레스트 모델 학습
rf_clf = RandomForestClassifier(n_jobs = -1, random_state = 156)
rf_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

pred = rf_clf.predict(X_test)

print('랜덤 포레스트 모델 이용 카테고리 분류결과:\n',
     classification_report(y_test, pred, target_names = df3.category.unique()))
```

```python
#나이브 베이즈 분류기 모델 학습
from sklearn.naive_bayes import MultinomialNB

nb_clf = MultinomialNB()
nb_clf.fit(X_train, y_train)
```

```python
pred = nb_clf.predict(X_test)

print('나이브 베이즈 분류기 이용 카테고리 분류결과:\n',
     classification_report(y_test, pred, target_names = df3.category.unique()))
```

## TF-IDf

df(t) = 특정 단어 t가 등장한 **문서 d의 개수**   


idf = df(t)의 반비례하는 수
- 즉 특정단어가 여러 문서에서 등장할수록 idf값은 낮아진다
## $idf(d, t) = ln(\frac{n}{1+df(t)})$
    -여기서 n은 전체 문서의 개수



## $idf(문서3, 노란) = ln(\frac{4}{1+1})$

## $=  0.693147$
